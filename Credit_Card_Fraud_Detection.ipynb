```python
#dependencies
import numpy as np #numerical works
import pandas as pd #its the data frame
from sklearn.model_selection import train_test_split #training and tesiting data
from sklearn.linear_model import LogisticRegression#use this model here
from sklearn.metrics import accuracy_score#performance of the model
```

```python
 #read the dataset to a Dataframe
df=pd.read_csv('/content/creditcard.csv')
```

```python
#first five rows
df.head()
```

```python
#last five rows
df.tail() 
```

```python
df.info()
```

```python
#checking num_of missing value in each column
df.isnull().sum()
```

```python
#distribution of legit transaction and fraudulent transactions
df['Class'].value_counts()
```

This dataset is a unbalanced 
0->normal transaction
1-fraudulent transaction

```python
#above unbalanced data because on set has a more values
#here i sperating the data for analysis
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#campare the values for both transactions
df.groupby('Class').mean()
```

Undr sampling dataset containing similar distributions of normal transactions and fradulant transactions


number of fraudulent transaction=88

```python
legit_sample=legit.sample(n=88)
```

 Concatinate two dataframes

```python
New_dataset=pd.concat([legit_sample,fraud],axis=0)#axis zero mean rows
```

```python
New_dataset.head()
```

```python
New_dataset.tail()
```

```python
New_dataset['Class'].value_counts()
```

```python
New_dataset.groupby('Class').mean()
```

Spiliting data into features & Targets

```python
X=New_dataset.drop(columns='Class',axis=1)
Y=New_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split data into training data& testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Training model


gentrally use binary classification problem use in logistic regression

```python
model=LogisticRegression()
```

```python
#training data use logistic regression
model.fit(X_train,Y_train)
```

Performance/Evloutions

Accuracy score

```python
#accuracy on training data
X_train_predict=model.predict(X_train)
tranining_accuracy=accuracy_score(X_train_predict,Y_train)
```

```python
print('The accuracy score on trainig data::',tranining_accuracy )
```

```python
#accuracy on test data
X_test_predict=model.predict(X_test)
testing_accuracy=accuracy_score(X_test_predict,Y_test)

```

```python
print('The accuracy score on testing data::',testing_accuracy)
```
